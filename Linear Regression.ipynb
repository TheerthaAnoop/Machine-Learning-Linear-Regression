```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
x_train=np.array([1.0,4.0])
y_train=np.array([300,400])
```

```python
m=len(x_train)
print(f"no.of training example is: {m}")
```

```python
i=1
x_i=x_train[i]
y_i=y_train[i]
print(f"x^({i}),y^({i}))=({x_i},{y_i})")
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title("housing prices")
plt.ylabel("Price(in 1000 dollers)")
plt.xlabel("Size(1000 sqft)")
plt.show()
```

```python
w=200
b=200
print(f"w: {w}")
print(f"b: {b}")
```

```python
def calculate_model_output(w,b,x):
    m=x.shape
    f_wb=np.zeros(m)
    for i in range(len(x)):
        f_wb[i]=w*x[i]+b
    return f_wb    
```

```python
tmp_f_wb=calculate_model_output(w,b,x_train)
plt.plot(x_train,tmp_f_wb,c="b",label="our prediction")
plt.scatter(x_train,y_train,marker="x",c="r",label="actual value")
plt.title("Housing price")
plt.ylabel("Price(in 1000 dollers)")
plt.xlabel("Size(1000 sqft)")
plt.legend()
plt.show()
```
